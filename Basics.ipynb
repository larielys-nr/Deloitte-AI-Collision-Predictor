```python
import numpy as np
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("car.data")
print(data.head())
```

```python
X_car = data[['buying', 'maint', 'safety']].values
y_car = data[['class']]

print(X_car, y_car)
```

```python
#as you can see above, the following data is in the format of strings and the ML algorithm cannot take in strings
#(in this case since the model is not trained for strings) so we need to pre-process the data to convert it to
#something the Algorithm will understand

#converting x
Le = LabelEncoder()
for i in range(len(X_car[0])):
    X_car[:, i] = Le.fit_transform(X_car[:, i])
    
print(X_car)
```

```python
#converting y
label_mapping = {
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}

y_car['class'].map(label_mapping)
```
