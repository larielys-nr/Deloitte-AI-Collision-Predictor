```python
import csv
from pathlib import Path
import pandas as pd

path_cwd = Path.cwd()
if Path.exists(path_cwd/"train_data.csv")==False:
    print("Shit")
else:
    print("Phew")
train_data_path = Path.cwd()/'train_data.csv'
print(train_data_path)

train_data = pd.read_csv(train_data_path)

import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('train_data.csv')

# Assuming the labels are in a column named 'label_column', replace it with the actual column name
labels = df['risk']

# If you just want the labels as a numpy array, you can use:
labels_array = labels.values

# Print the first few labels to verify
train_data = train_data.drop(columns=['c_object_type'])

print(labels.head())
print(train_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import accuracy_score

# Assuming you have gathered features (X) and labels (y) from your dataset

from sklearn.impute import SimpleImputer

# Create an imputer object with a strategy (e.g., mean, median, mode)
imputer = SimpleImputer(strategy='mean')

# Apply imputation to the DataFrame
train_data_imputed = imputer.fit_transform(train_data)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(train_data_imputed, labels_array, test_size=0.2, random_state=42)

# Create and train the Lasso model
lasso_model = Lasso(alpha=0.1)  # You can adjust the regularization parameter 'alpha' as needed
lasso_model.fit(X_train, y_train)

# Predict probabilities on the test set
y_pred_proba = lasso_model.predict(X_test)

# Convert probabilities to binary predictions (collision or no collision)
y_pred = (y_pred_proba > 0.5).astype(int)

# Evaluate the model
#accuracy = accuracy_score(y_test, y_pred)
#print("Accuracy:", accuracy)

from sklearn.metrics import mean_squared_error

# Convert continuous predictions to binary labels using a threshold (e.g., 0.5)
y_pred_binary = (y_pred > 0.5).astype(int)

# Evaluate the model using mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

```
