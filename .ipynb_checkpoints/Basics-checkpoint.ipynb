```python
"""
**************************************************
--------------------------------------------------
   NEAREST NEIGBHORS CLASSIFICATION MODEL (KNN)
--------------------------------------------------
**************************************************

Using the car.data

based on this youtube tutorial:
https://www.youtube.com/watch?v=pqNCD_5r0IU

from 25:45 to 44:00
"""
```

```python
import numpy as np
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("car.data")
print(data.head())
```

```python
X_car = data[['buying', 'maint', 'safety']].values
y_car = data[['class']]

print(X_car)
print(y_car)
```

```python
#as you can see above, the following data is in the format of strings and the ML algorithm cannot take in strings
#(in this case since the model is not trained for strings) so we need to pre-process the data to convert it to
#something the Algorithm will understand

#converting x
Le = LabelEncoder()
for i in range(len(X_car[0])):
    X_car[:, i] = Le.fit_transform(X_car[:, i])
    
print(X_car)
```

```python
#converting y
label_mapping = {
    'unacc':0,
    'acc': 1,
    'good': 2,
    'vgood': 3
}

y_car['class'].replace(label_mapping, inplace=True)
print(y_car)
```

```python
#create the model (a knn object)
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights='uniform')

#splitting data for training and testing
X_train_car, X_test_car, y_train_car, y_test_car = train_test_split(X_car, y_car, test_size=0.2)

print(X_car)
print(y_car)

#training the model
knn.fit(X_train_car, y_train_car.values.ravel())

#testing the prediction
prediction = knn.predict(X_test_car)

#calculating the accuracy
accuracy = metrics.accuracy_score(y_test_car, prediction)
print("predictions: ", prediction)
print("accuracy: ", accuracy)
```

```python
#Checking Actual and Predicted values (change the variable a to pick an index)

a = 1727
print("index: ", a)
print("actual value: ", y_car['class'][a])
print("predicted value: ", knn.predict(X_car)[a])
```

```python
"""
***********************
-----------------------
SUPPRT VECTOR MACHINES
-----------------------
***********************

SVMs basically help get a better boundary for classification.

Video:
https://www.youtube.com/watch?v=pqNCD_5r0IU

SVM Theory:
from 44:00 to 51:10

SVM Example:
from 51:11 to 57:45 
"""
```

```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()

#split it in features and labels

X_i = iris.data
y_i = iris.target


#all the data
print(X_i,y_i)
```

```python
#dimensions of the X (features) and Y (labels)
print(X_i.shape)
print(y_i.shape)

classes = ['Iris Setosa', 'Iris Versicolor', 'Iris Virginica']
```

```python
#hours of study vs good/bad grades
#10 students
#8 students -> Training Data
#2 studnets -> Testing Data (test out the model accuracy)

X_train, X_test, y_train, y_test = train_test_split(X_i, y_i, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#creating the model
model = svm.SVC()
model.fit(X_train, y_train)

print(model)

#testing the model
predictions = model.predict(X_test)
acc = accuracy_score(y_test, predictions)

print("predictions: ", predictions)
print("actual: ", y_test)
print(acc)

#note how the index of predictions gives a number which is used as the index for the classes list we defined
print(classes[predictions[0]])
print(classes[predictions[1]])
print(classes[predictions[5]])
```
